Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Import Dataset

```python
data = pd.read_csv('basic3.csv')
x = data.iloc[:, :-1].values
print(x)
```

Using the Elbow method to determine the number of Clusters

```python
from sklearn.cluster import KMeans

wcss = []

for i in range(1, 11):
    k_means = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    k_means.fit(x)
    wcss.append(k_means.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('WCSS Plot')
plt.xlabel('Clusters')
plt.ylabel('WCSS')
plt.show()
```

Train the K-Means Cluster on the data

```python
k_means = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)
y_group = k_means.fit_predict(x)
print(y_group)
```

Ploting the K_Means Clustering

```python
plt.scatter(x[y_group == 0, 0], x[y_group == 0, 1], s = 10, color = 'blue', label = 'Group 1')
plt.scatter(x[y_group == 1, 0], x[y_group == 1, 1], s = 10, color = 'green', label = 'Group 2')
plt.scatter(x[y_group == 2, 0], x[y_group == 2, 1], s = 10, color = 'red', label = 'Group 3')
plt.scatter(x[y_group == 3, 0], x[y_group == 3, 1], s = 10, color = 'pink', label = 'Group 4')
plt.scatter(k_means.cluster_centers_[:, 0], k_means.cluster_centers_[:, 1], color = 'black', s = 50)
plt.title('K_Means Cluster')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch

dendrogrm = sch.dendrogram(sch.linkage(x, method = 'ward'))
plt.title('dendrogram')
plt.xlabel('Basic3')
plt.ylabel('Eculidean Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hr = AgglomerativeClustering(n_clusters = 4, linkage = 'ward')
y_grp = hr.fit_predict(x)
print(y_grp)
```

```python
plt.scatter(x[y_grp == 0, 0], x[y_grp == 0, 1], s = 10, color = 'blue', label = 'Group 1')
plt.scatter(x[y_grp == 1, 0], x[y_grp == 1, 1], s = 10, color = 'green', label = 'Group 2')
plt.scatter(x[y_grp == 2, 0], x[y_grp == 2, 1], s = 10, color = 'red', label = 'Group 3')
plt.scatter(x[y_grp == 3, 0], x[y_grp == 3, 1], s = 10, color = 'pink', label = 'Group 4')
plt.title('Hierarchical Clustering')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
