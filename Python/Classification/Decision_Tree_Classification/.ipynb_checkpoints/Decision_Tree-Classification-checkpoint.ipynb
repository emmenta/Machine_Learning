Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Import Dataset

```python
data = pd.read_csv('Social_Network_Ads.csv')
print(data)
```

```python
x = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

Split the data into training and test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

Train the Decision Tree Classification on the training set

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
dtc.fit(x_train, y_train)
```

```python
print(dtc.predict(sc.transform([[30, 87000]])))
```

Predict the test set results

```python
y_pred = dtc.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```

Create a confusion matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

Visualizing the Training set results

```python
from matplotlib.colors import ListedColormap

# Visualize the data
x_set = sc.inverse_transform(x_train)
y_set = y_train
x1, x2 = np.meshgrid(
    np.arange(start = x_set[:, 0].min() - 10, stop = x_set[:, 0].max() + 10, step = 0.25),
    np.arange(start = x_set[:, 1].min() - 1000, stop = x_set[:, 1].max() + 1000, step = 0.25)
)

# Predict the set results
z = dtc.predict(sc.transform(np.array([x1.ravel(), x2.ravel()]).T)).reshape(x1.shape)

# Set the chat
plt.contourf(x1, x2, z, alpha = 0.75, cmap = ListedColormap(('skyblue', 'grey')))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())

# Plot the Decision Tree Classifier graph
Colors = ListedColormap(('blue', 'white'))

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1], color = Colors(i), label = j)

plt.title('Decision Tree Classifier (Training Set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
