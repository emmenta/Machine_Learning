IMPORT LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

IMPORT DATASET

```python
data = pd.read_csv('Data.csv')
print(data)
```

```python
x = data.iloc[:, :-1].values
print(x)
```

```python
y = data.iloc[:, -1].values
print(y)
```

SPLITING THE DATA INTO TRAINNG AND TEST SETS (TEST SIZE = 20%)

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 20)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

IMPLEMENT FEATURE SCALING

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
print(x_train)
```

LOGISTICS REGRESSION

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state = 0)
lr.fit(x_train, y_train)
```

```python
lr_pred = lr.predict(x_test)
```

K-NEAREST NEIGHBOR

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
knn.fit(x_train, y_train)
```

```python
knn_pred = knn.predict(x_test)
```

SUPPORT VECTOR CLASSIFICATION

```python
from sklearn.svm import SVC
svc = SVC(kernel = 'linear', random_state = 0)
svc.fit(x_train, y_train)
```

```python
svc_pred = svc.predict(x_test)
```

KERNEL SUPPORT VECTOR CLASSIFICATION

```python
from sklearn.svm import SVC
ksvc = SVC(kernel = 'rbf', random_state = 0)
ksvc.fit(x_train, y_train)
```

```python
ksvc_pred = ksvc.predict(x_test)
```

NAIVE BAYES CLASSIFICATION

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train, y_train)
```

```python
nb_pred = nb.predict(x_test)
```

DECISION TREE CLASSIFICATION

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
dtc.fit(x_train, y_train)
```

```python
dtc_pred = dtc.predict(x_test)
```

RANDOM FOREST CLASSIFICATION

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators = 40, criterion = 'entropy', random_state = 0)
rfc.fit(x_train, y_train)
```

```python
rfc_pred = rfc.predict(x_test)
```

CHECK THE ACCURACY SCORE OF EACH CLASSIFICATION

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print(
    'Accuracy score for Logistics Regression ', accuracy_score(y_test, lr_pred), '\n', confusion_matrix(y_test, lr_pred), '\n'
    'Accuracy score for K-Nearest Neghbors ', accuracy_score(y_test, knn_pred), '\n', confusion_matrix(y_test, knn_pred), '\n'
    'Accuracy score for Support Vector Classifier ', accuracy_score(y_test, svc_pred), '\n', confusion_matrix(y_test, svc_pred), '\n'
    'Accuracy score for Kernel Support Vector ', accuracy_score(y_test, ksvc_pred), '\n', confusion_matrix(y_test, ksvc_pred), '\n'
    'Accuracy score for Naive Bayes ', accuracy_score(y_test, nb_pred), '\n', confusion_matrix(y_test, nb_pred), '\n'
    'Accuracy score for Decision Tree ', accuracy_score(y_test, dtc_pred), '\n', confusion_matrix(y_test, dtc_pred), '\n'
    'Accuracy score for Random Forest', accuracy_score(y_test, rfc_pred), '\n', confusion_matrix(y_test, rfc_pred), '\n'
)
```
