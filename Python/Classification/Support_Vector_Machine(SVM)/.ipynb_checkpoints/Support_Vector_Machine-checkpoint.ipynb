Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Import dataset

```python
data = pd.read_csv('Social_Network_Ads.csv')
print(data)
```

```python
x = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

Splitting the dataset into training and test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

Train the SVM model on the training set

```python
from sklearn.svm import SVC
svc = SVC(kernel = 'linear', random_state = 0)
svc.fit(x_train, y_train)
```

Predicting a new result

```python
print(svc.predict(sc.transform([[30, 87000]])))
```

Prediciting the test set

```python
y_pred = svc.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```

Make a confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_pred, y_test)
print(cm)
accuracy_score(y_pred, y_test)
```

Visualizing the Training set result

```python
from matplotlib.colors import ListedColormap

# Visualize the training set result
x_set = sc.inverse_transform(x_train)
y_set = y_train
x1, x2 = np.meshgrid(
    np.arange(start = x_set[:, 0].min() - 10, stop = x_set[:, 0].max() + 10, step = 0.25),
    np.arange(start = x_set[:, 1].min() - 1000, stop = x_set[:, 1].max() + 1000, step = 0.25)
)

# Predict the class for each meshgrid
z = svc.predict(sc.transform(np.array([x1.ravel(), x2.ravel()]).T)).reshape(x1.shape)

# Create the chat
plt.contourf(x1, x2, z, alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())

# Plot the SVC graph for the training set
Colors = ListedColormap(('red', 'green'))
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1], color = Colors(i), label = j)
plt.title('Support Vector Classifier (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

Visualizing the Test set result

```python
# Visualizing the results of the test set
x_set = sc.inverse_transform(x_test)
y_set = y_test
x1, x2 = np.meshgrid(
    np.arange(start = x_set[:, 0].min() - 10, stop = x_set[:, 0].max() + 10, step = 0.25),
    np.arange(start = x_set[:, 1].min() - 1000, stop = x_set[:, 1].max() + 1000, step = 0.25)
)

# Predict the class for each meshgrid
z = svc.predict(sc.transform(np.array([x1.ravel(), x2.ravel()]).T)).reshape(x1.shape)

# Create the chat for the graph
plt.contourf(x1, x2, z, alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())

# Plot the SVC graph for the test set
Colors = ListedColormap(('red', 'green'))
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1], color = Colors(i), label = j)
plt.title('SUpport Vector Classifier (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
