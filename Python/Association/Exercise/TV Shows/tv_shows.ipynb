Import Libraries

```python
import numpy as np
import pandas as pd
```

Import Dataset

```python
data = pd.read_csv('TV Shows.csv', header = None)
print(data)
```

```python
transaction = []

for i in range(0, data.shape[0]):
    transaction.append([
        str(data.values[i, j]) for j in range(0, data.shape[1])
    ])
```

```python
from apyori import apriori
rules = apriori(transactions = transaction, min_support = 0.002, min_confidence = 0.3, min_lift = 2, min_length = 2, max_length = 2)
```

```python
results = list(rules)
results
```

```python
def inspect(results):
    movie1 = []
    movie2 = []
    support = []
    confidence = []
    lift = []

    for result in results:
        movie1.append(tuple(result[2][0][0])[0]) if result[2][0][0] else None
        movie2.append(tuple(result[2][0][1])[0]) if result[2][0][0] else None
        support.append(result[1])
        confidence.append(result[2][0][2])
        lift.append(result[2][0][3])

    return list(zip(movie1, movie2, support, confidence, lift))

resultinDF = pd.DataFrame(inspect(results), columns = ['Movie One', 'Movie Two', 'Supports', 'Confidence', 'Lifts'])
resultinDF
```

```python
resultinDF.nlargest(n = 15, columns = 'Lifts')
```
