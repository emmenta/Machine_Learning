Import Libraries

```python
import numpy as np
import pandas as pd
```

Import Dataset

```python
data = pd.read_csv('basket.csv', header = None)
print(data)
```

```python
transaction = []

for i in range(0, data.shape[0]):
    transaction.append([
        str(data.values[i, j]) for j in range(0, data.shape[1])
    ])
```

Train the Apriori Model on the data

```python
from apyori import apriori
rules = apriori(transactions = transaction, min_support = 0.001, min_confidence = 0.3, min_lift = 1, min_length = 2, max_length = 2)
```

Visualize the data as of the Rules

```python
results = list(rules)
results
```

Display the data into a Data Frame

```python
def inspect(results):
    prd1 = []
    prd2 = []
    support = []
    confidence = []
    lift = []

    for result in results:
        prd1.append(tuple(result[2][0][0])[0]) if result[2][0][0] else None
        prd2.append(tuple(result[2][0][1])[0]) if result[2][0][1] else None
        support.append(result[1])
        confidence.append(result[2][0][2])
        lift.append(result[2][0][3])

    return list(zip(prd1, prd2, support, confidence, lift))


resultinDF = pd.DataFrame(inspect(results), columns = ['Product 1', 'Product 2', 'Support', 'Confidence', 'Lifts'])
resultinDF
```

Sort the data in descending order of Support

```python
resultinDF.nlargest(n = 15, columns = 'Lifts')
```

```python
resultinDF.nlargest(n = 15, columns = 'Support')
```
