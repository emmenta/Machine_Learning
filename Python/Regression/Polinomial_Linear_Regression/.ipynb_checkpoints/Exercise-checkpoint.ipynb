Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Create the table

```python
data = {
    "conc" : [0.07,0.09,0.08,0.16,0.17,0.21,0.49,0.58,0.53,1.22,1.15,1.07,2.84,2.57,3.10],
    "time" : [9,9,9,7,7,7,5,5,5,3,3,3,1,1,1]
}
data = pd.DataFrame(data)
print(data)
```

```python
x = data.iloc[:, 1].values
x = x.reshape(len(x), 1)
print(x)
```

```python
y = data.iloc[:, :-1].values
print(y)
```

Plot the Linear Regression

```python
from sklearn.linear_model import LinearRegression
slr = LinearRegression()
slr.fit(x, y)
```

```python
plt.scatter(x, y, color = 'blue')
plt.plot(x, slr.predict(x), color = 'red')
plt.title('Concenation vs Time')
plt.xlabel('time')
plt.ylabel('conc')
plt.show()
```

```python
print(slr.predict([[3.5]]))
```

Do the Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poli = PolynomialFeatures(degree = 4)
x_poli = poli.fit_transform(x)
plr = LinearRegression()
plr.fit(x_poli, y)
```

Plot the Polynomial Regression

```python
plt.scatter(x, y, color = 'blue')
plt.plot(x, plr.predict(x_poli), color = 'red')
plt.title('Concenation vs Time')
plt.xlabel('time')
plt.ylabel('conc')
plt.show()
```

```python
x_grid = np.arange(min(x), max(x), 0.1)
x_grid = x_grid.reshape(len(x_grid), 1)
plt.scatter(x, y, color = 'blue')
plt.plot(x_grid, plr.predict(poli.fit_transform(x_grid)), color = 'red')
plt.title('Concenation vs Time')
plt.xlabel('time')
plt.ylabel('conc')
plt.show()
```

```python
print(plr.predict(poli.fit_transform([[3.5]])))
```

```python
print(plr.predict(poli.fit_transform([[1.2]])))
```
