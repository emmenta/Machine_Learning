Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Import Dataset

```python
data = pd.read_csv('university.csv')
print(data.head(10))
```

```python
x = data.iloc[:, 3].values
y = data.iloc[:, 1].values
```

Split the data into training and test sets

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .2, random_state = 24)
x_train = x_train.reshape(-1,1)
```

Train the model for SLR

```python
from sklearn.linear_model import LinearRegression
slr = LinearRegression()
slr.fit(x_train, y_train)
```

Compare the Results of the test and predicted results by ploting a graph

```python
y_pred = slr.predict(x_test.reshape(-1,1))
```

```python
np.set_printoptions(precision = 2)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```

```python
plt.scatter(x_train, y_train, color = 'red')
plt.plot(x_train, slr.predict(x_train), color = 'blue')
plt.title('Regression plot for ACT and GPA (Training set)')
plt.xlabel('ACT')
plt.ylabel('GPA')
plt.show()
```

```python
plt.scatter(x_test, y_test, color = 'blue')
plt.plot(x_train, slr.predict(x_train), color = 'red')
plt.title('Regression plot for ACT and GPA (Test set)')
plt.xlabel('ACT')
plt.ylabel('GPA')
plt.show()
```
