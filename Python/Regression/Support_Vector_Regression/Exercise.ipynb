Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Create table

```python
data = {
    "conc": [0.07,0.09,0.08,0.16,0.17,0.21,0.49,0.58,0.53,1.22,1.15,1.07,2.84,2.57,3.10],
    "time": [9,9,9,7,7,7,5,5,5,3,3,3,1,1,1]
}
data = pd.DataFrame(data)
print(data)
```

```python
x = data.iloc[:, 1].values
x = x.reshape(len(x), 1)
print(x)
```

```python
y = data.iloc[:, :-1].values
print(y)
```

Scale X and Y variables

```python
from sklearn.preprocessing import StandardScaler
x_sc = StandardScaler()
y_sc = StandardScaler()
x = x_sc.fit_transform(x)
y = y_sc.fit_transform(y)
```

```python
print(x)
```

```python
print(y)
```

Train the SVR model

```python
from sklearn.svm import SVR
svr = SVR(kernel = 'rbf')
svr.fit(x, y)
```

Plot the SVR graph

```python
x_grid = np.arange(min(x), max(x), 0.1)
x_grid = x_grid.reshape(len(x_grid), 1)
plt.scatter(x_sc.inverse_transform(x), y_sc.inverse_transform(y), color = 'red')
plt.plot(x_sc.inverse_transform(x_grid), y_sc.inverse_transform(svr.predict(x_grid).reshape(-1,1)), color = 'blue')
plt.title('Concentration vs Time')
plt.xlabel('Time')
plt.ylabel('Conc')
plt.plot()
```

```python
print(y_sc.inverse_transform(svr.predict(x_sc.transform([[3.5]])).reshape(-1,1)))
```
